# Breakeven Analysis

```python
import numpy as np # from numpy import arange, ceil
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Function that takes 
def breakeven(fixed_cost, # Fixed Cost associated with manufacturing or sourcing a product
              unit_sp, # Selling price per unit of the product
              unit_vp, # Variable cost per unit for the product
              exp_sales, # Expected number of units that will be sold
              new_price = None # The price if there is a price change
             ):
    # Generate the number of units from 0 to expected numbers of units that will be sold with step of 100
    quantity = np.arange(0, exp_sales, 100)
    
    # Calculate the breakeven point for the product at the selling price
    bep_quantity_1 = fixed_cost/(unit_sp - unit_vp)
     
    # Total cost associated with the product
    total_cost = fixed_cost + (unit_vp * quantity)
    
    # Total Revenue for the unit selling price
    total_rev = quantity * unit_sp
    
    # Profit at the unit selling price
    profit = total_rev - total_cost
    
    res = {'Break Even Point' : np.ceil(bep_quantity_1)}
    
    if new_price is not None:
        # Calculate the breakeven point for the product at a different selling price
        bep_quantity_2 = fixed_cost/(new_price - unit_vp)

        # Total Revenue at the unit selling price at the new price
        total_rev_2 = quantity * new_price

        # Profit at the new selling price
        profit_2 = total_rev_2 - total_cost
        
        res = {'Break Even Point' : np.ceil(bep_quantity_1), 'Break Even Point for Price Match' : np.ceil(bep_quantity_2)}
    
    # Create subplots to plot Revenue and profit in two different plots
    fig = make_subplots(rows=2, shared_xaxes = True, x_title = "Number of Units", column_titles=["Breakeven Analysis"],
                       y_title = "Total Cost")
    
    # Plot Total Cost Vs. Quantity 
    fig.add_trace(go.Scatter(x=quantity, y=total_cost, name = "Total Cost"), row=1, col=1)
    
    # Add Total Revenue to the first plot
    fig.append_trace(go.Scatter(x=quantity, y=total_rev, name = "Total Revenue"), row=1, col=1)
    
    if new_price is not None:
        # Add Total Revenue when the product is sold at a different price
        fig.append_trace(go.Scatter(x=quantity, y=total_rev_2, name = "Revenue for Price Match"), row=1, col=1)

    # Plot Total Cost Vs. Quantity 
    fig.add_trace(go.Scatter(x=quantity, y=total_cost,name = "Total Cost"), row=2, col=1) 
    
    # Add Profit to the second plot
    fig.append_trace(go.Scatter(x=quantity, y=profit, name = "Profit"), row=2, col=1)
    
    if new_price is not None:
        # Add Profit when the product is sold at the new selling price
        fig.append_trace(go.Scatter(x=quantity, y=profit_2, name = "Profit for Price Match"), row=2, col=1)
    
    # Ensure when hover a straight line is displayed and shows the value of the coordinates
    fig.update_traces(mode="markers+lines", hovertemplate=None)   
    fig.update_layout(hovermode="x unified")        
               
    
    return fig.show(), res
        
```
